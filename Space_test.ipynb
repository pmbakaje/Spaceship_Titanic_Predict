```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#pd.pandas.set_option('display.max_columns',None)
```

```python
df_test=pd.read_csv('test.csv')
```

```python
df_test.shape
```

```python
df_test.isnull().sum()
```

```python
Numerical_feature=[feature for feature in df_test.columns if df_test[feature].dtype!='O' and df_test[feature].dtype!='bool']
```

```python
Numerical_feature
```

```python
Categorical_feature=[feature for feature in df_test.columns if df_test[feature].dtype=='O']
```

```python
Categorical_feature
```

```python
for feature  in Categorical_feature:    
    df_test[feature].fillna(df_test[feature].mode()[0], inplace=True)
```

```python
df_test.isnull().sum()
```

```python
df_test.duplicated().sum()
```

```python
for feature in Numerical_feature:
    df_test[feature].fillna(df_test[feature].median(),inplace=True)
```

```python
#df_test['Transported']=df_test['Transported'].astype(int)
df_test['VIP']=df_test['VIP'].astype(int)
df_test['CryoSleep']=df_test['CryoSleep'].astype(int)
```
